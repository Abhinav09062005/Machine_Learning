```python
import numpy as np
import pandas as pd
```

```python
date = pd.read_csv('/kaggle/input/19560-indian-takeaway-orders/restaurant-1-orders.csv')
```

```python
date.head()
```

```python
date.info()
```

```python
# Converting to datetime datatype
date['Order Date'] = pd.to_datetime(date['Order Date'])
```

```python
#this is first way to remove time from date column
date['Order Date'] = pd.to_datetime(date['Order Date'], dayfirst=True).dt.strftime('%d/%m/%Y')
```

```python
date.info()
```

```python
#this is second way to remove time from date column
date['Order Date'] = pd.to_datetime(date['Order Date'], dayfirst=True).dt.date
```

```python
date.info()
```

```python
# Converting to datetime datatype
date['Order Date'] = pd.to_datetime(date['Order Date'])
```

```python
date.info()
```

```python
#extract year
date['date_year'] = date['Order Date'].dt.year

date.sample(5)
```

```python
#extract month
date['date_month_no'] = date['Order Date'].dt.month

date.head()
```

```python
date['date_month_name'] = date['Order Date'].dt.month_name()

date.head()
```

```python
#extract day
date['date_day'] = date['Order Date'].dt.day

date.head()
```

```python
# day of week
date['date_dow'] = date['Order Date'].dt.dayofweek

date.head()
```

```python
 #day of week - name

date['date_dow_name'] = date['Order Date'].dt.day_name()

date.drop(columns=['Order Number','Item Name']).head()
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['Order Number','Item Name']).head()
```

```python
#Extract week of the year
date['date_week'] = date['Order Date'].dt.week
date.drop(columns=['Order Number','Item Name']).head()
```

```python
date['quarter'] = date['Order Date'].dt.quarter
date.drop(columns=['Order Number','Item Name']).head()
```

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)
date.drop(columns=['Order Number','Item Name']).head()
```
